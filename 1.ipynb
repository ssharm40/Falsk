## Flask
is a python web application framework.
With help of flask you can create rest api that will be usefull in soft. dev
Flask is based on WSGI(Web Server Gateway Interface) and jinja2 template.


### WSGI
It's a technique that describes how a server communicate with web applications
and how web application can be chained together to process one request.

### jinja2:
It's a templae for programming purpose so that you can change the pagesn runtime, or you will be able to use multiple kind of programming syntax in flask

1)   {%
       It's used for statement
     %}


2)   {{
       It's used fopr expression to print the template output
     }}


3)   {#
       For comment
     #}

4)   #..## => used for line statements

Exp:
{%for todo in alltodo%}
 <tr>
	<th scope="row">{{loop.index}}</th>
	<td>{{todo.title}} </td>    
 </tr>
{%endfor%}












### Significance of G in Python Flask:
G stands for global.
That data saved in G lost after the context ends, and it's not an appopriate place to store data between requests.
